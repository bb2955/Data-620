```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
parks = [
    "Magic Kingdom",
    "EPCOT",
    "Hollywood Studios",
    "Animal Kingdom",
    "Disneyland Park",
    "California Adventure"
]

G.add_nodes_from(parks)
```

```python
edges = [
    ("Magic Kingdom", "EPCOT"),
    ("Magic Kingdom", "Hollywood Studios"),
    ("Magic Kingdom", "Animal Kingdom"),
    ("EPCOT", "Hollywood Studios"),
    ("Hollywood Studios", "Animal Kingdom"),
    ("Disneyland Park", "California Adventure")
]

G.add_edges_from(edges)
```

```python
plt.figure(figsize=(10, 7))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="skyblue",
    node_size=3000,
    font_size=10,
    font_weight="bold",
    edge_color="gray"
)

plt.title("Hello, Graph World: Disney Parks Network", fontsize=14)
plt.show()
```

```python
print("Nodes:")
print(G.nodes())

print("\nEdges:")
print(G.edges())
```
